```python
from pyspark.sql import DataFrame
from pyspark.sql.functions import col, current_timestamp, lit
```

```python
def add_audit_columns(df: DataFrame) -> DataFrame:
    """
    データフレームに監査用カラム（処理日時、ソース）を追加する関数
    """
    return df.withColumn("processed_at", current_timestamp()) \
             .withColumn("source_system", lit("Databricks_Sample"))

def filter_expensive_data(df: DataFrame, threshold: int = 100) -> DataFrame:
    """
    指定した価格以上のデータを抽出する関数
    """
    return df.filter(col("price") >= threshold)
```
